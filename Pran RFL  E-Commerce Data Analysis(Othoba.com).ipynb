```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
###Loading E-commerce Data (2023 Jan - 2024 Sep) for Analysis
df = pd.read_csv(r'D:\My projects\Pran RFL E-commerce Data (2023 Jan - 2024 Sep).csv')
```

```python
###Checking for Missing Values in the E-commerce Dataset
print(df.isnull().sum())
```

```python
###Removing Unnecessary Columns from the E-commerce Dataset
df.drop(columns=['Product Attribute', 'OrderShippingExclTax', 'Supplier'], inplace=True)
```

```python
print(df.columns)
```

```python
###Filling Missing Values in 'Order Cancel Reason' Column with 'Successful Order'
df['Order Cancel Reason'].fillna('Successful Order', inplace= True)
```

```python
###Converting Date Columns to Datetime Format
df['Order CreatedOn'] = pd.to_datetime(df['Order CreatedOn'], errors= 'coerce')
df['ShippedDate'] = pd.to_datetime(df['ShippedDate'], errors='coerce')
df['DeliveredDate'] = pd.to_datetime(df['DeliveredDate'], errors= 'coerce')


```

# Explanatory 360Â° Analysis of Othoba's Monthly Sales for the Year 2023-2024


```python
### Yearly Order Trends Analysis (2023 vs 2024)

order_counts = df.groupby(['FinalOrderStatus', 'Year']).size().unstack(fill_value=0)
order_counts.plot(kind='bar', figsize=(12, 6), colormap='Paired', rot=45)
plt.title('Yearly Order: 2023 vs 2024', fontsize=16)
plt.xlabel('Order Status', fontsize=12)
plt.ylabel('Number of Orders', fontsize=14)
plt.legend(title='Year')

for bar in plt.gca().containers:
    plt.bar_label(bar, fmt='%d', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
### Category-wise Order Performance Analysis

category_sales = df['SubOrderId'].nunique()
category_sales = df.groupby('Category')['SubOrderId'].nunique().sort_values(ascending=False)
plt.figure(figsize=(20,10))
sns.set_style=("ticks")
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Category Order Count', fontsize=18, fontweight='bold')
plt.xlabel('Category', fontsize=15)
plt.ylabel('Unique Orders', fontsize=15)
plt.xticks(rotation=35, ha ='right', fontsize=12, fontweight='bold')

for index, value in enumerate(category_sales.values):
    plt.text(index, value + (value * 0.02), f'{value:,.0f}', ha='center' , fontsize=15, fontweight='bold')

plt.legend([],[], frameon=False)
plt.tight_layout()
plt.show()





```

```python
### Division-wise Order Distribution

division_order_counts = df['ShippingDivision'].value_counts()
total_orders_division = division_order_counts.sum()

plt.figure(figsize=(15,10))
sns.barplot(x=division_order_counts.index, y=division_order_counts.values,palette='BrBG')
plt.title('Division wise Order', fontsize=16, fontweight='bold')
plt.xlabel('Shipping Division', fontsize=12, fontweight='bold')
plt.ylabel('Number of Orders', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=15, fontweight='bold')
for i, v in enumerate(division_order_counts.values):
    percentage = (v/total_orders_division) * 100
    plt.text(i, v + 10, f'{v}\n({percentage:.2f}%)', ha='center', fontsize=15, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
### State-wise Order Distribution

state_order_counts = df['ShippingStateProvince'].value_counts()
total_orders_state = division_order_counts.sum()

plt.figure(figsize=(17,8))
sns.barplot(x=state_order_counts.head(15).index, y=state_order_counts.head(15).values,palette='CMRmap')
plt.title('Top 10 Shipping State Order Count', fontsize=16, fontweight='bold')
plt.xlabel('Shipping State', fontsize=12, fontweight='bold')
plt.ylabel('Number of Orders', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10, fontweight='bold')
for i, v in enumerate(state_order_counts.head(15).values):
    percentage = (v/total_orders_state) * 100
    plt.text(i, v + 10, f'{v}\n({percentage:.2f}%)', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
####Sales and Cancellations by Division (Heatmap Analysis)
division_order_count = df.groupby(['ShippingDivision','FinalOrderStatus']).size().unstack(fill_value=0)

top_10_divisions = division_order_count.sum(axis=1).nlargest(10).index
division_order_counts_top10 = division_order_count.loc[top_10_divisions]

plt.figure(figsize=(15,8))
sns.heatmap(division_order_counts_top10, annot=True, fmt='d', cmap='RdYlGn', linewidths=0.5, linecolor='black')
plt.title('Divison wise Sales & Cancel', fontsize=18, fontweight='bold')
plt.xlabel('Final Order Status', fontsize=18, fontweight='bold')
plt.ylabel('Divisions',fontsize=18, fontweight='bold')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()


```

```python
###Shipment Method Performance Analysis

shipment_method_count=df['ShipmentBy'].value_counts().head(12)
percentages=(shipment_method_count/shipment_method_count.sum()) * 100

plt.figure(figsize=(12,8))
ax= sns.barplot(y=shipment_method_count.index, x=shipment_method_count.values, palette='RdBu')

plt.title('Top 12 Shipment Methods by Order Count', fontsize=18, fontweight='bold')
plt.xlabel('Count', fontsize=14, fontweight='bold')
plt.ylabel('Shipment Method', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
sns.despine(left=True, bottom=True)
for i , count in enumerate(shipment_method_count):
    percentage = percentages[i]
    ax.text(count, i , f'{count}({percentage:.2f}%)', ha = 'left', va= 'center', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
### Total Revenue Analysis by Year
total_sell_yearly = df.groupby('Year')['TotalSellingPrice'].sum()

plt.figure(figsize=(8,5))
ax = total_sell_yearly.plot(kind='bar', color=['blue','orange'])

for i, value in enumerate(total_sell_yearly):
    ax.text(i, value + (value * 0.01), f'{value:,.0f}', ha='center', fontsize=12, fontweight='bold')

plt.title("Total Sell by Year", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=14, fontweight='bold')
plt.ylabel("Total Profit", fontsize=14, fontweight='bold')
plt.xticks(rotation=0, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')


plt.tight_layout()

plt.show()
```

```python
### Filling REF! value with Corporate
df['B2C/B2B'] = df['B2C/B2B'].replace('#REF!', 'Corporate')
```

```python
### B2B vs B2C Sales Contribution

b2b_b2c_sell = df.groupby('B2C/B2B')['TotalSellingPrice2'].sum()

colors = ['#4C72B0', '#C44E52']
explode = (0.05, 0.05)
plt.figure(figsize=(7, 7))
wedges, texts, autotexts = plt.pie(
    b2b_b2c_sell,
    labels=b2b_b2c_sell.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=explode,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
    textprops={'fontsize': 12, 'fontweight': 'bold'}
)
for i, text in enumerate(autotexts):
    text.set_text(f"{b2b_b2c_sell.iloc[i]:,.0f} BDT\n({text.get_text()})")
    text.set_color('white')
    text.set_fontsize(13)
    text.set_fontweight('bold')

plt.title("B2B/B2C Sales", fontsize=16, fontweight='bold')
plt.ylabel("")

plt.show()
```

```python
### Vendor Performance Analysis (Top & Low Performers)

top_vendors = df.groupby('Vendor Group').agg({'TotalSellingPrice2': 'sum'}).reset_index()
top_vendors = top_vendors.sort_values(by = 'TotalSellingPrice2', ascending=False).head(10)

plt.figure(figsize=(19,6))
sns.barplot  (x='TotalSellingPrice2', y='Vendor Group', data=top_vendors, palette='Set2')
plt.title('Top Vendor  by Sell', fontsize=16)
plt.xlabel('Total revenue BDT', fontsize=12)
plt.ylabel('Category', fontsize=12)

for i, v in enumerate(top_vendors['TotalSellingPrice2']):
    plt.text(v + 0.01 * max(top_vendors['TotalSellingPrice2']), i, f'{v:.2f} BDT', ha='left', va='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

```

```python
###Division wise total sell

division_profit = df.groupby('ShippingDivision')['TotalSellingPrice2'].sum().sort_values(ascending=False)
total_profit = division_profit.sum()

plt.figure(figsize=(18, 13))
sns.barplot(x=division_profit.head(10).index, y=division_profit.head(10).values, palette='magma')

plt.title('Total Sell Shipping Division Top 10', fontsize=16, fontweight='bold')
plt.xlabel('Shipping Division', fontsize=12, fontweight='bold')
plt.ylabel('Total Sell', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=12, fontweight='bold')

for i, v in enumerate(division_profit.head(10).values):
    percentage = (v / total_profit) * 100
    plt.text(i, v + (v * 0.01), f'{v}\n{percentage:.2f}%', ha='center', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
### Pricing Comparison (Regular Price, Selling Price, and Discount) for 2023 vs. 2024
total_regular_price_2023 = df[df['Year'] == 2023]['RegularPrice'].sum()
total_regular_price_2024 = df[df['Year'] == 2024]['RegularPrice'].sum()

total_selling_price_2023 = df[df['Year'] == 2023]['TotalSellingPrice'].sum()
total_selling_price_2024 = df[df['Year'] == 2024]['TotalSellingPrice'].sum()

total_discount_2023 = total_regular_price_2023 - total_selling_price_2023
total_discount_2024 = total_regular_price_2024 - total_selling_price_2024

price_summary = pd.DataFrame({
    'Year': ['2023', '2024'],
    'Total Regular Price': [total_regular_price_2023, total_regular_price_2024],
    'Total Selling Price': [total_selling_price_2023, total_selling_price_2024],
    'Total Discount': [total_discount_2023, total_discount_2024]
}).set_index('Year')

plt.figure(figsize=(10, 6))
sns.heatmap(price_summary, annot=True, fmt=".2f", cmap="RdYlBu", linewidths=0.5, linecolor='black')
plt.title("Pricing Comparison: 2023 vs 2024", fontsize=16, fontweight='bold')
plt.xlabel("Category", fontsize=13, fontweight='bold')
plt.ylabel("Year", fontsize=14, fontweight='bold')

plt.show()
```

```python
###early Financial Summary: Total Revenue, Cost, and Profit for 2023 and 2024
df_2023 = df[df['Year'] == 2023]
df_2024 = df[df['Year'] == 2024]
yearly_summary = pd.DataFrame({
    'Year': ['2023', '2024'],
    'Total Revenue': [df_2023['TotalSellingPrice'].sum(), df_2024['TotalSellingPrice'].sum()],
    'Total Cost': [df_2023['TotalCostPrice'].sum(), df_2024['TotalCostPrice'].sum()],
    'Total Profit': [df_2023['Profit'].sum(), df_2024['Profit'].sum()]
}).set_index('Year')
plt.figure(figsize=(10, 6))

sns.heatmap(yearly_summary, annot=True, fmt=".2f", cmap="RdYlGn", linewidths=0.5, linecolor='black')
plt.title("Pricing Comparison: 2023 vs 2024", fontsize=16, fontweight='bold')
plt.xlabel("Category", fontsize=13, fontweight='bold')
plt.ylabel("Year", fontsize=14, fontweight='bold')

plt.show()
```

```python
#Worst 5 Categories by Revenue (Top Categories with the Lowest Total Selling Price)
import squarify
division_profit = df.groupby('Category')['TotalSellingPrice2'].sum().sort_values(ascending=True)

plt.figure(figsize=(12, 8))
squarify.plot(sizes=division_profit.head(5), label=division_profit.head(5).index, alpha=.8, color=sns.color_palette('YlGnBu'))
plt.title('Worst 5 Category by Revenue', fontsize=16, fontweight='bold')
plt.axis('off')

plt.show()
```

```python
### Top 10 Reasons for Order Cancellations and Their Percentages

df2 = df[df['FinalOrderStatus'] == 'Cancel']
cancel_reason_counts = df2['Order Cancel Reason'].value_counts()
total_cancellations = cancel_reason_counts.sum()
top_10_cancellations = cancel_reason_counts.head(10)
top_10_percentages = (top_10_cancellations / total_cancellations) * 100

plt.figure(figsize=(20, 7))
sns.barplot(y=top_10_cancellations.index, x=top_10_cancellations.values, palette='coolwarm')

plt.title('Top 10 Reasons for Order Cancellations', fontsize=16, fontweight='bold')
plt.xlabel('Number of Cancellations', fontsize=14, fontweight='bold')
plt.ylabel('Cancellation Reason', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

for i, (v, p) in enumerate(zip(top_10_cancellations.values, top_10_percentages.values)):
    plt.text(v + 0.5, i, f'{v} ({p:.2f}%)', va='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
###Preferred Payment Methods for Successful Orders
successful_orders = df[df['FinalOrderStatus'] == 'Sales']
payment_method_counts = successful_orders['Payment Method'].value_counts()
colors = sns.color_palette("husl", len(payment_method_counts))
plt.figure(figsize=(12, 6))
bars = plt.barh(payment_method_counts.index, payment_method_counts.values, color=colors, edgecolor="black")
for bar, count in zip(bars, payment_method_counts.values):
    plt.text(count + 50, bar.get_y() + bar.get_height()/2, f'{count:,}', va='center', fontsize=12, fontweight='bold')

plt.title('Preferred Payment Methods (Successful Orders)', fontsize=16, fontweight='bold', color='#333')
plt.xlabel('Number of Successful Orders', fontsize=12, fontweight='bold', color='#555')
plt.ylabel('Payment Method', fontsize=12, fontweight='bold', color='#555')

plt.tight_layout()
plt.show()
```

```python
### On-time vs Late Deliveries
df['DeliveryTime'] = (df['DeliveredDate'] - df['ShippedDate']).dt.days
df['DeliveryStatus'] = df['DeliveryTime'].apply(lambda x: 'On Time' if x <= 5 else 'Late')
delivery_counts = df['DeliveryStatus'].value_counts(normalize=True) * 100

plt.figure(figsize=(6, 6))
delivery_counts.plot(kind='pie', autopct='%1.1f%%', colors=['green', 'red'])
plt.title("On-time vs Late Deliveries")
plt.ylabel("")
plt.show()
```

```python
### Order Cancellations Due to Shipping Delays
canceled_due_to_shipping = df[df['Order Cancel Reason'].str.contains("delay", na=False, case=False)]

plt.figure(figsize=(8, 5))
sns.countplot(y=canceled_due_to_shipping['Order Cancel Reason'], order=canceled_due_to_shipping['Order Cancel Reason'].value_counts().index, palette='coolwarm')
plt.xlabel("Number of Orders Canceled")
plt.ylabel("Cancellation Reason")
plt.title("Order Cancellations Due to Shipping Delays")
plt.show()
```

```python
###Vendors with the Highest Delivery Delays

df2 = df[df['FinalOrderStatus'] == 'Sales']
vendor_delays = df2.groupby('Category')['DeliveryStatus'].value_counts(normalize=True).unstack().fillna(0)
vendor_delays = vendor_delays.sort_values(by='Late', ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=vendor_delays.index, y=vendor_delays['Late'] * 100, palette='Reds_r')
plt.xticks(rotation=90)
plt.ylabel("Late Delivery Percentage")
plt.xlabel("Vendor Group")
plt.title("Vendors with Highest Delivery Delays")
plt.show()
```

```python
###Top 10 Most Profitable Categories

df2 = df[df['FinalOrderStatus'] == 'Sales']
category_profit = df2.groupby('Category')[['TotalSellingPrice', 'Profit']].sum().sort_values(by='Profit', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(y=category_profit.index[:10], x=category_profit['Profit'][:10], palette='Blues_r')
plt.xlabel("Total Profit")
plt.ylabel("Category")
plt.title("Top 10 Most Profitable Categories")
plt.show()
```

```python
###Top 10 Loss-Making Vendors

df['ProfitMargin'] = (df['Profit'] / df['TotalSellingPrice']) * 100
loss_making_vendors = df[df['Profit'] < 0].groupby('Vendor')['Profit'].sum().sort_values()
plt.figure(figsize=(10, 5))
sns.barplot(y=loss_making_vendors.index[:10], x=loss_making_vendors[:10], palette='Reds_r')
plt.xlabel("Total Loss")
plt.ylabel("Vendor")
plt.title("Top 10 Loss-Making Vendors")
plt.show()
```

```python
###Order Cancellation Rate by Payment Method
df2 = df[df['FinalOrderStatus'] == 'Cancel']
cancelled_orders = df2[df['FinalOrderStatus'] == 'Cancel']
payment_cancellations = cancelled_orders['Payment Method'].value_counts(normalize=True) * 100
plt.figure(figsize=(20, 8))
sns.barplot(x=payment_cancellations.index, y=payment_cancellations.values, palette='Reds_r')
plt.xlabel("Payment Method", fontsize=14, fontweight='bold')
plt.ylabel("Cancellation Rate (%)", fontsize=14, fontweight='bold')
plt.title("Order Cancellation Rate by Payment Method", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')

plt.show()
```

```python
### COD vs. Online Payments: Order Status Breakdown
cod_vs_others_count = df.groupby('Payment Method')['FinalOrderStatus'].value_counts().unstack().fillna(0)
cod_vs_others_percent = cod_vs_others_count.div(cod_vs_others_count.sum(axis=1), axis=0) * 100
fig, ax = plt.subplots(figsize=(12, 6))
cod_vs_others_percent.plot(kind='bar', stacked=True, colormap='coolwarm', ax=ax)


plt.xlabel("Payment Method", fontsize=12)
plt.ylabel("Percentage", fontsize=12)
plt.title("COD vs. Online Payments", fontsize=14)
plt.xticks(rotation=30, ha='right', fontsize=10) 
plt.yticks(fontsize=10)
plt.legend(title="Order Status", fontsize=10)


plt.tight_layout()  
plt.show()
```

```python
###Delivery Durations (Days) in 2024

df = df[df['Order CreatedOn'].dt.year == 2024]
df['DeliveryDuration'] = (df['DeliveredDate'] - df['Order CreatedOn']).dt.days
df = df[(df['DeliveryDuration'] >= 1) & (df['DeliveryDuration'] <= 40)]
plt.figure(figsize=(20, 8))
sns.histplot(df['DeliveryDuration'], bins=30, kde=True, color='blue')
plt.title('Delivery Durations (Days) - 2024', fontsize=16)
plt.xlabel('Delivery Duration (Days)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)


for patch in plt.gca().patches:
    height = patch.get_height()
    if height > 0:
        plt.gca().annotate(f'{int(height)}', xy=(patch.get_x() + patch.get_width() / 2, height),
                           xytext=(0, 5), textcoords='offset points', ha='center', fontsize=10, color='black')
plt.show()
```
